# 성별추이
- 막대그래프(BAR CHART)
- 선 도표(LINE CHART)
- 면적그래프(AREA CHART)
- 원그래프/파이도(PIE CHART)

```python
# 월별 성별추이
# 막대그래프(BAR CHART)
df[['MALE','FEMALE']].plot(kind='bar')
plt.show()
```

```python
# 월별 성별추이
# 선 도표(LINE CHART)
df[['MALE','FEMALE']].plot(kind='line')
plt.show()
```

```python
# 월별 성별추이
# 면적그래프(AREA CHART)
df[['MALE','FEMALE']].plot(kind='area')
plt.show()
```

```python
lido_GENDER = df.iloc[:,:2]
print(lido_GENDER)
lido_GENDER = lido_GENDER.T
lido_GENDER.T
```

```python
# 월별 성별추이
# 원그래프/파이도(PIE CHART)
lido_GENDER.plot.pie(subplots=True, figsize=(27, 27))
```
